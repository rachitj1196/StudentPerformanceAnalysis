## Data Cleaning

```python
#check for null values
data.isnull().sum()
```

```python
#dropping the redundant data now which are not so important for data analysis
student=data.drop(['race/ethnicity', 'parental level of education'], axis=1)
```

```python
student.head()
```

## relationship analysis

```python
#correlation matrix - to summarize data for advanced analysis
corel=student.corr()
```

```python
#using heatmap to show the data using seaborn library
sns.heatmap(corel,xticklabels=corel.columns, yticklabels=corel.columns, annot=True)
```

```python
#To visualize relationship between 2 variables
sns.pairplot(student)
```

```python
#Scatter Plot
sns.scatterplot(x='math score', y='writing score', hue='gender', data=student)
```

```python
sns.relplot(x='math score', y='reading score', hue='lunch', data=student)
#hue is used to group variables to produce points with different colors (here- male, female)
```

```python
# to get idea of shape and spread of data, we use Histogram
sns.distplot(student['math score'], bins=5)
```

```python
#categorical Plot
sns.catplot(x='reading score', kind='box', data=student)
```
